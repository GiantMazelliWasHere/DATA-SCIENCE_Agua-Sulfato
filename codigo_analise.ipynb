```python
import pandas as pd
import statistics as stats
```

```python
# Função que calcula a moda de um dataframe

def moda(df):
    for i in df.iterrows():
        return stats.mode(i[1][1:])
```

```python
# Importando os arquivos

agua = pd.read_excel('dataframes_originais/agua.xlsx')
sulfato = pd.read_excel('dataframes_originais/sulfato.xlsx')
```

# Dataframe com os dados sobre à água

```python
# Criando o dataframe com os dados

agua_df = pd.DataFrame(agua)
```

```python
# Retirando a coluna ID/Água do dataframe para facilitar os calculos

agua_df_numeros = agua_df.drop(columns=['ID/Água'])
```

```python
# Media, mediana e moda dos dados de agua

agua_df['Média'] = agua_df_numeros.mean(axis=1)
agua_df['Mediana'] = agua_df_numeros.median(axis=1)
agua_df['Moda'] = moda(agua_df)
```

```python
# Desvio médio dos dados de agua

agua_df['Desvio Médio Pt'] = abs(agua_df_numeros['Pt'] - agua_df['Média'])
agua_df['Desvio Médio P0,5m'] = abs(agua_df_numeros['P0,5m'] - agua_df['Média'])
agua_df['Desvio Médio P1,5m'] = abs(agua_df_numeros['P1,5m'] - agua_df['Média'])
agua_df['Desvio Médio P2,5m'] = abs(agua_df_numeros['P2,5m'] - agua_df['Média'])
agua_df['Desvio Médio P3,5m'] = abs(agua_df_numeros['P3,5m'] - agua_df['Média'])
```

```python
# Variância, Desvio Padrão e Coeficiente de Variação dos dados de agua

agua_df['Variância'] = agua_df_numeros.var(axis=1)
agua_df['Desvio Padrão'] = agua_df_numeros.std(axis=1)
agua_df['Coeficiente de Variação'] = (agua_df['Desvio Padrão'] / agua_df['Média']) * 100
```

```python
# Print final do dataframe de agua

agua_df
```

```python
# Aredondando os valores para 2 casas decimais

agua_df_arredondado = agua_df.round(2)

agua_df_arredondado
```

```python
# Passando o dataframe para um arquivo excel

agua_df.to_excel('dataframes_finais/agua_final.xlsx')
agua_df_arredondado.to_excel('dataframes_finais/agua_final_arredondado.xlsx')
```
